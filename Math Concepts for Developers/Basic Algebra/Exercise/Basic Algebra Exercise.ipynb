### Problem 1. Polynomial Interpolation
We know that if we have a set of $n$ data points with coordinates $(x_1; y_1), (x_2; y_2), \dots, (x_n; y_n)$, we can try to figure out what function may have generated these points.

Please note that **our assumptions about the data** will lead us to choosing one function over another. This means that our results are as good as our data and assumptions. Therefore, it's extremely important that we write down our assumptions (which sometimes can be difficult as we sometimes don't realize we're making them). It will be better for our readers if they know what those assumptions and models are.

In this case, we'll state two assumptions:
1. The points in our dataset are generated by a polynomial function
2. The points are very precise, there is absolutely no error in them. This means that the function should pass **through every point**

This method is called *polynomial interpolation* (*"polynomial"* captures assumption 1 and *"interpolation"* captures assumption 2).

It can be proved (look at [Wikipedia](https://en.wikipedia.org/wiki/Polynomial_interpolation) for example) that if we have $n$ data points, there is only one polynomial of degree $n-1$ which passes through them. In "math speak": "the vector spaces of $n$ points and polynomials of degree $n-1$ are isomorphic (there exists a bijection mapping one to the other)".

There are a lot of ways to do interpolation. We can also write the function ourselves if we want but this requires quite a lot more knowledge than we already covered in this course. So we'll use a function which does this for us. `numpy.polyfit()` is one such function. It accepts three main parameters (there are others as well, but they are optional): a list of $x$ coordinates, a list of $y$ coordinates, and a polynomial degree.

Let's say we have these points:
```python
points = np.array([(0, 0), (1, 0.8), (2, 0.9), (3, 0.1), (4, -0.8), (5, -1.0)])
```

First, we need to "extract" the coordinates:
```python

y = points[:, 1]
```

Then, we need to calculate the interpolating polynomial. For the degree, we'll initially set $n-1$:
```python

# TODO: Find a way to make the line shorter; call the correct arguments
```

After that, we need to plot the function. To do this, we'll create a range of $x$ values and evaluate the polynomial at each value:
```python

```

Finally, we need to plot the result. We'll plot both the fitting polynomial curve (using `plt.plot()`) and the points (using `plt.scatter`). It's also nice to have different colors to make the line stand out from the points.

````python
points = np.array([(0, 0), (1, 0.8), (2, 0.9), (3, 0.1), (4, -0.8), (5, -1.0)])

xs = points[:, 0]
ys = points[:, 1]

polynomial = poly.Polynomial.fit(xs, ys, len(points) - 1)

plot_x = np.linspace(np.min(xs), np.max(xs), 1000)
plot_y = polynomial(plot_x)

plt.plot(plot_x, plot_y, c = "cyan")
plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

**Be extremely careful to ensure that the function uses its parameters!** Of course, you can extract other utility functions if you wish (e.g., separating plotting from the rest is a good idea).

We'll use this function to try some other cases.

````python
def interpolate_polynomial(points, degree, min_x, max_x):
    """
    Interpolates a polynomial of the specified degree through the given points and plots it
    points - a list of points (x, y) to plot
    degree - the polynomial degree
    min_x, max_x - range of x values used to plot the interpolating polynomial
    """

    xs = points[:, 0]
    ys = points[:, 1]

    polynomial = poly.Polynomial.fit(xs, ys, degree)
    
    plot_x = np.linspace(min_x, max_x)
    plot_y = polynomial(plot_x)

    plt.scatter(xs, ys)
    plt.xlabel("x")
    plt.xlabel("y")
    plt.plot(plot_x, plot_y)
    plt.show()
````

````python
points = np.array([(0, 0), (1, 0.8), (2, 0.9), (3, 0.1), (4, -0.8), (5, -1.0)])
interpolate_polynomial(points, len(points) - 1, np.min(points[:, 0]), np.max(points[:, 0]))
````

### Problem 3. Complex Numbers as Vectors
We saw that a complex number $z = a + bi$ is equivalent to (and therefore can be represented as) the ordered tuple $(a; b)$, which can be plotted in a 2D space.
````python
def plot_complex_number(z):
    """
    Plots the complex number z as a radius vector in the 2D space
    """

    currentAxis = plt.gca()

    currentAxis.spines["bottom"].set_position("zero")
    currentAxis.spines["top"].set_visible(False)
    currentAxis.spines["left"].set_position("zero")
    currentAxis.spines["right"].set_visible(False)

    plt.quiver(0, 0, z.real, z.imag, angles = "xy", scale_units = "xy", scale = 1)
    plt.xticks(range(-4, 4))
    plt.yticks(range(-4, 4))

plot_complex_number(2 + 3j)
````

How about many numbers? We'll need to get a little bit more creative. First, we need to create a 2D array, each element of which will be a 4-element array: `[0, 0, z.real, z.imag]`. Next, `plt.quiver()` can accept a range of values. Look at [this StackOverflow post](https://stackoverflow.com/questions/12265234/how-to-plot-2d-math-vectors-with-matplotlib) for details and adapt your code.

````python
def plot_complex_numbers(numbers, colors):
    """
    Plots the given complex numbers as radius vectors in the 2D space
    """

    currentAxis = plt.gca()

    currentAxis.spines["bottom"].set_position("zero")
    currentAxis.spines["top"].set_visible(False)
    currentAxis.spines["left"].set_position("zero")
    currentAxis.spines["right"].set_visible(False)

    real_parts = [num.real for num in numbers]
    imag_parts = [num.imag for num in numbers]
    zeros = [0 for num in numbers]

    plt.quiver(zeros, zeros, real_parts, imag_parts, angles = "xy", scale_units = "xy", scale = 1, color = colors)

    # Changed in order to be able to see the whole vectors
    plt.xticks(range(-6, 6))
    plt.yticks(range(-6, 6))
    
plot_complex_numbers([2 + 3j, -2 - 1j, -3, 2j], ["green", "red", "blue", "orange"])
````

### Problem 4. Recursion and Fractals
<img src="recursion.jpg" style="max-height: 230px" />

> "To understand recursion, you first need to understand recursion."

There are three main parts to a recursive function:
1. Bottom - when the recursion should finish
2. Operation - some meaningful thing to do
3. Recursive call - calling the same function
4. Clean-up - returning all data to its previous state (this reverses the effect of the operation)

Let's do one of the most famous recursion examples. The most trivial function to recreate is the **factorial**. Derive (or look up) a recurrent definition of the factorial of a non-negative integer and implement it.

**There is a one-to-one correspondence between loops and recursion.** Use this correspondence to create an iterative definition (one which uses loops) and compare both. Test that they return the same results.

**Note:** Think how Python handles very large integers.

````python
def factorial_recursive(n):
    """
    Computes the factorial of a non-negative number n (using recursion)
    """

    if (n == 1):
        return 1
    
    return n * factorial_recursive(n-1)

def factorial_iterative(n):
    """
    Computes the factorial of a non-negative number n (using iteration)
    """
    output = 1
    for i in range (1, n + 1):
        output *= i

    return output
````

Test our functions:

````python
print(factorial_recursive(5))
print(factorial_iterative(5))

print(factorial_iterative(100) == factorial_recursive(100))
````

#### Endless sequences
A very famous recursive definition gives us natural numbers (by (the Peano axioms)[https://en.wikipedia.org/wiki/Peano_axioms]). How can we recreate this using Python? Obviously, we cannot put an infinite sequence in a list. However, we can do something else: define the entire sequence and take only parts of it.

Devise a way to define all natural numbers, using the function signature below. Use this to define a smaller, but just as infinite, sequence of even numbers. **Hint:** If you're very stuck, think about the problem hard and it will finally _yield_ to your efforts.

````python
def nat():
    """Returns all natural numbers"""
    n = 0
    while True:
        yield n
        n += 1
def even():
    """Returns all even numbers"""
    n = 0
    while True:
        yield n
        n += 2
````

Use the approach above to rewrite the factorial function in a "lazy" way.

````python
def factorial_new(n):
    """Returns the factorial numbers from 0! to n! lazily"""
    fact = 1
    yield fact  # 0!
    for i in range(1, n + 1):
        fact *= i
        yield fact
````

Use **the same approach** to return the Fibonacci sequence - another famous one in mathematics. Given `n`, produce all Fibonacci numbers from $F_1$ up to $F_n$, inclusive.

````python
def fibonacci(n):
    """Returns the Fibonacci sequence lazily"""
    if n == 1:
        yield 1
    else:
        yield 1  
        yield 1  
        fib_1, fib_2 = 1, 1
        
        for i in range(2, n):
            fib_1, fib_2 = fib_2, fib_1 + fib_2
            yield fib_2
````

````python
def mandelbrot_set_recursively(element_index, constant):
        """ Function to recursively get the current element in the Mandelbrot set """
        if element_index == 0:
             return 0
        else:
             return mandelbrot_set_recursively(element_index - 1, constant) ** 2 + constant
````

Iterative Approach:

````python
def mandelbrot_set_iteratively(c):
        """ Function to iteratively get the current element in the Mandelbrot set """
        z = 0
        while True:
            yield z
            z = z ** 2 + c
````

The below functions are used to create and display the Mandelbrot sets:

````python
def complex_matrix(xmin, xmax, ymin, ymax, pixel_density):
    """ Function to create a combination of complex numbers in a fixed rectangle (xmin, xmax, ymin, ymax) """
    re = np.linspace(xmin, xmax, int((xmax - xmin) * pixel_density))
    im = np.linspace(ymin, ymax, int((ymax - ymin) * pixel_density))
    return re[np.newaxis, :] + im[:, np.newaxis] * 1j
````

````python
def is_stable(c, num_iterations):
    """ Function to conclude whether an element c is present in the Mandelbrot set """
    z = 0
    for _ in range(num_iterations):
        z = z ** 2 + c
    return abs(z) <= 2
````

````python
def get_members(c, num_iterations):
    """ Function to filter the members of the matrix so that only valid ones are present """
    mask = is_stable(c, num_iterations)
    return c[mask]
````

````python
def graph_mandelbrot_set(xmin, xmax, ymin, ymax, pixel_density):
    """ Function to graph the Mandelbrot square depending on the rectangle of coordinates and pixel density  """
    matrix = complex_matrix(xmin, xmax, ymin, ymax, pixel_density=pixel_density)
    members = get_members(matrix, num_iterations=20)
    plt.scatter(members.real, members.imag, color="black", marker=",", s=1)
    plt.gca().set_aspect("equal")
    plt.axis("off")
    plt.tight_layout()
    plt.show()
````

````python
def l_system(rules, axiom, iterations):
    """ Function to iterate a given axiom and transform it using the system rules """
    for _ in range(iterations):
        axiom = l_translation(rules, axiom)
    return axiom
````

### Problem 5. Run-length Encoding
One application of algebra and basic math can be **compression**. This is a way to save data in less space than it originally takes. The most basic form of compression is called [run-length encoding](https://en.wikipedia.org/wiki/Run-length_encoding).

Write a function that encodes a given text. Write another one that decodes.

We can see that RLE is not very useful in the general case. But it can be extremely useful if we have very few symbols. An example of this can be DNA and protein sequences. DNA code, for example, has only 4 characters.

Test your encoding and decoding functions on a DNA sequence (you can look up some on the Internet). Measure how much your data is compressed relative to the original.

````python
def encode(text):
    """
    Returns the run-length encoded version of the text
    (numbers after symbols, length = 1 is skipped)
    """
    if not text:
        return ""
    
    encoded_text = ""
    current_char = text[0]
    current_count = 1

    for char in text[1:]:
        if char == current_char:
            current_count += 1
        else:
            if current_count > 1:
                encoded_text += f"{current_char}{current_count}"
            else:
                encoded_text += current_char

            current_char = char
            current_count = 1

    if current_count > 1:
        encoded_text += f"{current_char}{current_count}"
    else:
        encoded_text += current_char

    return encoded_text

def decode(text):
    """
    Decodes the text using run-length encoding
    """
    decoded_text = ""

    for char in text:
        if char.isdigit():
            decoded_text += decoded_text[-1] * (int(char) - 1)
        else:
            decoded_text += char
    
    return decoded_text
````

Plot all multiplication and factorization times as a function of the number of bits. You should see that factorization is much, much slower. If you don't see this, just try larger numbers :D.

````python
def generate_random_number(bits):
    return secrets.randbits(bits)

def measure_times(bits, repetitions):
    factorization_times = []
    multiplication_times = []

    for _ in range(repetitions):
        number = generate_random_number(bits)

        start_time = time.time()
        factors = factorint(number)
        factorization_times.append(time.time() - start_time)

        start_time = time.time()
        product = 1
        
        for factor, count in factors.items():
            product *= factor ** count
            
        multiplication_times.append(time.time() - start_time)

    avg_factorization_time = sum(factorization_times) / repetitions
    avg_multiplication_time = sum(multiplication_times) / repetitions

    return avg_factorization_time, avg_multiplication_time
````

````python
def plot_factorization_times(bits, repetitions):
    factorization_results = []
    multiplication_results = []
    
    for b in bits:
        factorization_time, multiplication_time = measure_times(b, repetitions)
        factorization_results.append(factorization_time)
        multiplication_results.append(multiplication_time)
    
    plt.plot(bits, factorization_results, label="Factorization Time", marker='o')
    plt.plot(bits, multiplication_results, label="Multiplication Time", marker='o')
    plt.xlabel('Bit Length')
    plt.ylabel('Time (seconds)')
    plt.title('Factorization vs Multiplication Time')
    plt.legend()
    plt.grid(True)
    plt.show()
````

````python
def diffie_hellman_simulation(bits, base):
    p = generate_large_prime(bits)
    g = base

    a = secrets.randbelow(p - 2) + 2  
    A = pow(g, a, p)

    b = secrets.randbelow(p - 2) + 2
    B = pow(g, b, p)

    shared_secret_1 = pow(B, a, p)

    shared_secret_2 = pow(A, b, p)

    return shared_secret_1, shared_secret_2, p, g, A, B
````

### 1. Fields

In mathematics, a field is a set F on which operations addition, subtraction (addition of negative numbers), multiplication and division (multiplication by reciprocal number) are defined - generally, multiplication and addition are defined. A field must contain at least two elements. The axioms must be valid for the set:

Addition:
1. Commutativity:        $ x + y = y + x $
2. Associativity:        $ (x + y) + z = x + (y + z) $
3. Additive Identity:    $ \exists\, 0 \in F $ such that $ x + 0 = x $
4. Additive Inverse:     $ \forall x \in F, \exists\, -x \in F $ such that $ x + (-x) = 0 $

Multiplication:
1. Commutativity:        $ x·y = y·x $
2. Associativity:        $ (x·y)·z = x·(y·z) $
3. Multiplicative Identity: $ \exists 1 \in F, 1 ≠ 0 $ such that $ x·1 = x $
4. Multiplicative Inverse: $ \forall x \neq 0 \in F, \exists\, x^{-1} $ such that $ x·x^{-1} = 1 $

Distributivity:
1. Distributive Law:    $  x·(y + z) = x·y + x·z $ 

Example fields: set of real numbers - $ \mathbb{R} $

### 6. Perfect Secrecy

Perfect Secrecy is one of the most fundamental concepts of cryptography. In simple words, it means even if an attacker has infinite computation power, they cannot guess or obtain any information about the plain text (the normal text we convert / encode) or secrets from cipher text (encrypted text) - so learning the cipher text gives you no advantage in guessing the message. It was first proposed by Claude Shannon in 1946. If both the two participants share a secret key, they can encrypt it securely. They are the only ones who know and access the message. Even if an eavesdropper knows the cipher text, they learn nothing about the original message. So the cipher text reveals zero information about the original one.

### 7. Symmetric encryption

Having talked about perfect secrecy, we can talk about symmetric encryption - when the same secret key is used
for both encrypting and decrypting a message. Examples include AES (Advanced Encryption Standard), DES (Data Encryption Standard), etc. The hardest part of the procedure is when the two parties need to agree on a secret key. When they have agreed upon a key the encryption, when used, is very fast.

### 8. Encryption of 1-bit messages

Encryption of 1-bit messages include using XOR. When you want to encrypt a message "m" to obtain cipher text "c", you use $ \oplus $ (XOR) operation with a key "r" (random, known only to the sender and receiver)

Encryption algorithm: $$ c = m \oplus k $$ 

Decryption algorithm: $$ m = c \oplus k $$

Important! Key must be secret, random and used only once.

### 9. Encryption of multi-bit messages

In order to encrypt / decrypting multi-bit messages, we can extend our 1-bit encrypting to another level. For example, we could use the one-time pad encryption technique (OTP) - symmetric encryption technique. It cannot be cracked. A key is used (pre-shared) - its length is bigger or equal to the size of the message we are sending. Each bit of the plain text is encrypted by combining it with the corresponding bit from the "pad" (the key) - with modular addition or XOR $ (\oplus) $. The key must be truly random, it must not be reused in whole or in part and must be completely secret from other parties.

c - cipher text, m - plain text, k - key, n - length of plain text

$$ \textbf{Encryption Algorithm: } \\
\text{For each } i \text{ from } 0 \text{ to } n - 1: \\
c[i] = m[i] \oplus k[i] $$

$$ \textbf{Decryption Algorithm: } \\
\text{For each } i \text{ from } 0 \text{ to } n - 1: \\
m[i] = c[i] \oplus k[i] $$

This algorithm achieves perfect secrecy unless the key is reused.


### 10. Participants and why is One-Time Pad (OTP) decryptable

The system of one-time pad is decryptable because XOR is its own inverse! The special property of XOR validates this:

$$ a \oplus b \oplus b = a \quad \Rightarrow \quad (a \oplus b) \oplus b = a $$
thus if we encrypt a message "m" with key "k", we get the cipher text:
$ c = m \oplus k $, then, to decrypt, we just do:
$$ c \oplus k = (m \oplus k) \oplus k = m $$

Participants:

1. Party1
    - has the message "m"
	- has the shared secret key "k"
	- encrypts "m" - $ c = m \oplus k $
	- sends it to Party2
2. Party2
    - receives the message "c"
	- has the same shared secret key "k"
	- decrypts "c" - $ m = c \oplus k = (m \oplus k) \oplus k $
3. EavesDropper
    - sees "c", but cannot know "k"
	- if "k" is truly random and used once, the EavesDropper has
	no useful information to decrypt "c"

So, in this way, the eavesdropper would not be able to decrypt the cipher text - with no value for "k", there are milion possibilities for the key - milion possibilities for the message which are equally likely.

### 11. Reusing one-time pad

If we reuse the one-time pad, that means, we reuse the same key for encrypting multiple messages, in this way, we violate the perfect secrecy and there comes the possibility of attacks. For example:

If we have two messages:
m1 = "...."
m2 = ",,,,"

Their encrypted messages would be:
c1 = m1 XOR k
c2 = m2 XOR k

An eavesdropper sees c1 and c2 and tries to attack:
$$ c1 \oplus c2 = (m1 \oplus k) \oplus (m2 \oplus k) $$
$$ = m1 \oplus m2 \oplus k \oplus k $$
$$ = m1 \oplus m2 \oplus 0 $$
$$ = m1 \oplus m2 $$

Now the eavesdropper has $ m1 \oplus m2 $. Having the XOR of the two texts is dangerous! If the attacker guesses a part of the message, they can guess the leftone. This is because natural language has patterns (word order, punctuation, grammar, etc.) Having all things considered, a one-time pad must be used $ \textbf{ONLY ONCE} $.

### 13. Making our own little cryptosystem

````python
def xor_strings(s1, s2):
    """
        XOR two strings of the same length.
    """
    return ''.join(chr(ord(a) ^ ord(b)) for a, b in zip(s1, s2))

def encrypt(plaintext, key):
    """
        Encrypt the plaintext with the secret key.
    """
    return xor_strings(plaintext, key)

def decrypt(ciphertext, key):
    """
        Decrypt the ciphertext with the secret key (same as encryption).
    """
    return xor_strings(ciphertext, key)
````

## Huffman Coding, Compression and Entropy

### Abstract

This article explores various types of data compression, focusing in-depth on the Huffman coding algorithm. We also compare it with other compression algorithms such as LZ77. Additionally, we introduce the concept of entropy and its role in entropy-based compression.

### 1. Compression

There are two main types of compression: lossy and lossless.

a) Lossy Algorithms

Lossy compression reduces file size by permanently removing some of the original data. It is used when certain data can be discarded in exchange for significantly reduced storage size. For example, when compressing an image using a lossy algorithm, the algorithm removes details that are less noticeable to the human eye.

![Lossy](https://bitmovin.com/wp-content/uploads/2020/03/LossyCompressionDoggo.jpeg)

b) Lossless Algorithms

Lossless compression reduces file size by eliminating unnecessary metadata and redundancies without losing any actual data. For instance, in image compression, lossless methods remove non-essential metadata while preserving visual quality. The original file can be perfectly reconstructed after decompression.

![Lossless](https://cyberhoot.com/wp-content/uploads/2022/03/A-Complete-Guide-To-An-Image-Compression-1.jpeg)

c) Key Differences

The key difference is that lossy compression irreversibly removes data, whereas lossless compression preserves all original information. Lossless compression is preferred when accuracy is critical (e.g., text or data files), while lossy compression is more suitable for media such as images, audio and video where a certain amount of quality can be sacrificed for space savings.
    
    - Lossy methods offer greater file size reduction but may reduce quality.
    - Lossless methods maintain full quality but achieve smaller reductions in size.

![Comparison](https://www.adobe.com/uk/creativecloud/photography/discover/media_1847e38787b8377123e92dc4b1806580b39c732de.png?width=750&format=png&optimize=medium)

### 2. Entropy

Entropy is a measure of unpredictability in a system. It was introduced by Claude Shannon in 1948.

$$ H(X) = - \sum_{x \in X} p(x) \log_2 p(x) $$

In the context of data compression, entropy helps determine the minimum number of bits required to represent data without losing information. It forms the basis of entropy encoding, a category of lossless compression. Common examples include Huffman coding and arithmetic coding. If the entropy is higher, more bits are required. The goal of compression is to reduce entropy by eliminating redundancy in the data.

### 3. Huffman Coding
Huffman coding is a type of lossless entropy-based data compression technique. It was developed by David Huffman in 1952.

Given a set of symbols $ \text{S} $, and for each symbol $ \text{x} \in \text{S} $, a frequency 
$ \text{f}_x $ (representing how often the symbol appears in the data), the algorithm builds a binary tree where:

    - More frequent symbols receive shorter codes.
    - Less frequent symbols receive longer codes.

#### Steps of the algorithm:
1) Calculate the frequency of each character.
2) Create a leaf node for each symbol and insert them into a priority queue (min-heap), where lower frequency nodes have higher priority.
3) Build the tree:
    - Remove the two nodes with the lowest frequency.
    - Create a new node with these two as children; its frequency is their sum.
    - Insert the new node back into the queue.
4) Repeat step 3 until only one node remains. This becomes the root of the Huffman tree.
5) Assign codes by traversing the tree:
    - Go left: append 0.
    - Go right: append 1.
    - Each leaf's code is the path from the root to that leaf.

#### Example 1
Frequencies: A – 5, B – 9, C – 12, D – 13, E – 16, F – 45

Construction:

    - A + B = Node1 (14)
    
    - C + D = Node2 (25)
    
    - Node1 + E = Node3 (30)
    
    - Node2 + Node3 = Node4 (55)
    
    - Node4 + F = Root (100)

Resulting nodes:

F: 0
C: 100
D: 101
A: 1100
B: 1101
E: 111

#### Example 2
Frequencies: X – 1, Y – 1, Z – 1, W – 1

Steps:

    - X + Y = A (2)

    - Z + W = B (2)
    
    - A + B = Root (4

Resulting nodes:

X: 00
Y: 01
Z: 10
W: 11

### 4. Decompressing Huffman Codes

Compressed data is stored as a bitstring. For example, the string "XYXZ" becomes "00010010".

To decompress:

    - Use the Huffman tree and read the bitstring one bit at a time.

    - Traverse from the root: 0 - left, 1 - right.

    - When a leaf node is reached, output the symbol and return to the root.

    - Repeat until all bits are decoded.

Important: Huffman codes are prefix-free—no code is a prefix of another—ensuring the algorithm is unambiguous.

### 5. Storing the Huffman Tree

There are two main ways to store the Huffman tree:

    - Store the entire tree structure in the compressed file.

    - Store only the frequency table (e.g., in the file header) and reconstruct the tree during decompression.

The tree or frequency data is stored before the actual compressed content.

### 6. Implementation and Testing

````python
# Class for one Node in Huffman tree
class Node:
    def __init__(self, char, freq):
        """ 
        Constructor for Node class, setting the character and its frequency
        """
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

# Function for frequencies
def get_frequencies(text):
    """ 
    Function to calculate frequencies in a given string
    """
    freq = {}
    for char in text:
        if char not in freq:
            freq[char] = 0
        freq[char] += 1
    return freq

# Function for building the tree
def build_tree(freq_dict):
    """ 
    Function to build tree according to frequencies dictionary
    """
    nodes = [Node(ch, fr) for ch, fr in freq_dict.items()]

    while len(nodes) > 1:
        nodes.sort(key=lambda x: x.freq)
        
        left = nodes.pop(0)
        right = nodes.pop(0)

        merged = Node(None, left.freq + right.freq)
        merged.left = left
        merged.right = right

        nodes.append(merged)

    return nodes[0] 

# Function for generating codes
def generate_codes(node, current_code="", codes=None):
    """
    Function to traverse the tree and set codes
    """
    if codes is None:
        codes = {}
    if node.char is not None:
        codes[node.char] = current_code
    else:
        generate_codes(node.left, current_code + "0", codes)
        generate_codes(node.right, current_code + "1", codes)
        
    return codes

# Function for compressing the text
def compress(text, codes):
    return ''.join(codes[ch] for ch in text)

# Function for decompressing the text
def decompress(encoded, root):
    result = ""
    node = root
    for bit in encoded:
        if bit == '0':
            node = node.left
        else:
            
            node = node.right
            
        if node.char is not None:
            result += node.char
            node = root
    return result
````

````python
text = "AAAAAABBBBCCCCCDDEEEE"

freq = get_frequencies(text)
root = build_tree(freq)
codes = generate_codes(root)

print("Codes:")
for ch in codes:
    print(f"{ch}: {codes[ch]}")

compressed = compress(text, codes)
print("Compressed:", compressed)

decompressed = decompress(compressed, root)
print("Decompressed:", decompressed)

original_size = len(text) * 8 
compressed_size = len(compressed)

compression_ratio = (1 - (compressed_size / original_size)) * 100

print()

print(f"Original size: {original_size} bits")
print(f"Compressed size: {compressed_size} bits")
print(f"Compression ratio: {compression_ratio:.2f}%")
````

## Error Detection and Correction in Data Transmission

### Abstract
Data is often corruped by errors due to noise. These errors can lead to lower communication reliability and system performance. To address this, various error detection and correction techniques are implemented. This article explores transmission errors, their types, and the methods used to detect and correct them. Detailed explanations are provided with code examples to demonstrate concepts such as parity checks, checksums, and Cyclic Redundancy Check (CRC).

### 1. Errors

Errors occur when the received information differs from what was sent due to issues like noise during transmission. For example, a bit may flip from 0 to 1 or vice versa. Error-detection codes are used to identify and possibly correct such issues.

### 2. Types of errors

A) Single-Bit Error

Occurs when only one bit in the data is altered.

Example:

    Sent: 0011
    
    Received: 0111 (bit 2 changed)

![Single-bit](https://th.bing.com/th/id/R.730d2875fd82ec613b5fe0c7368bbe97?rik=ZeZv39mJ1UQoMg&pid=ImgRaw&r=0&sres=1&sresct=1)

B) Multiple-Bit Error

More than one bit is changed, although not necessarily consecutively.

Example:

    Sent: 0011
    
    Received: 0110 (bits 2 and 4 changed)

![Multiple-bit](https://image5.slideserve.com/9548647/multiple-bit-error-l.jpg)

c) Burst Error

Several consecutive bits are altered.

Example:

    Sent: 00111000
    
    Received: 01001000 (bits 2 to 4 changed)

![Burst](https://scaler.com/topics/images/burst-bit-error.webp)

### 3. Error Detection

To ensure data is transmitted accurately, error detection techniques add redundancy (extra bits) to the data. These help identify if any error has occurred during transmission.

### 4. Techniques in Error Detection

a) Simple Parity Check

Adds one bit to make the total number of 1s even (even parity) or odd (odd parity). This can detect single-bit errors.

    Example:

        Original: 1011 - Add 1 - 10111

![SP1](https://i.ytimg.com/vi/NaZ9aAut6Rw/maxresdefault.jpg)



![SP2](https://media.geeksforgeeks.org/wp-content/uploads/detect12.jpg)

b) Two-Dimensional Parity Check

Organizes data in a matrix and performs parity checks on rows and columns. It detects all single-bit and some multiple-bit errors.

    Example: 

        Original: 
        1011
        0100
        1110
        0001

Parity bits are added to rows and columns to catch errors.

![2DPC](https://image1.slideserve.com/2243303/two-dimensional-parity-check-l.jpg)

c) Checksum

Sums segments of data using 1’s complement arithmetic and sends the checksum along with the data. If the sum doesn’t match, an error is detected.

![CS](https://media.geeksforgeeks.org/wp-content/uploads/detect13.jpg)

d) Cyclic Redundancy Check (CRC)

Uses polynomial division to calculate a CRC remainder, which is appended to the message. The receiver performs the same division and checks for a remainder of zero.

    Example:

        Data: 11010110

        Generator: 1011

        Transmitted: 11010110100

![CRC1](https://media.geeksforgeeks.org/wp-content/uploads/20240528161237/Binary-Division-Rules.webp)

![CRC2](https://media.geeksforgeeks.org/wp-content/uploads/detect14.jpg)

### 5. Error Correction

Once an error is detected, it can be corrected in two ways:

    Forward Error Correction (FEC): Receiver corrects the error without requesting retransmission.

    Backward Error Correction: Sender resends the data if an error is detected.

### 6. Bit Modifications

Types of bit errors that may occur:

    Single-Bit Error: One bit is flipped.

    Burst Error: Several bits in sequence are changed.

    Multiple-Bit Error: Several non-consecutive bits are altered.

    Bit Deletion: One bit is missing in the received data.

    Bit Insertion: An extra bit is added.

### 7. Needed Redundant Bits

Redundancy must be optimized to balance between accuracy and bandwidth efficiency.

### 8. Practical Examples

````python
# a) Even Parity in Python

def parity_bit(data):
    """ 
    Function for calculating parity bit
    """
    ones_count = bin(data).count('1')

    if ones_count % 2 == 0:
        return 0
    else:
        return 1
````

````python
data = 0b101010
parity = parity_bit(data)
print(f"Data: {bin(data)}, Parity Bit: {parity}")
````

````python
# b) 2D Parity in Python
def two_d_parity(matrix):
    """ 
    Function for calculating the 2D parity
    """
    rows, cols = matrix.shape
    row_parity = np.sum(matrix, axis=1) % 2
    col_parity = np.sum(matrix, axis=0) % 2
    
    matrix_with_row_parity = np.column_stack((matrix, row_parity))
    
    matrix_with_parity = np.row_stack((matrix_with_row_parity, np.append(col_parity, 0)))
    
    return matrix_with_parity
````

````python
matrix = np.array([[1, 0, 1], [0, 1, 1], [1, 1, 0]])
parity_matrix = two_d_parity(matrix)
print(f"Original Matrix:\n{matrix}")
print(f"Parity Matrix:\n{parity_matrix}")
````

````python
def crc(data, polynomial=0x11d, size=8):
    """ 
    Function for evaluating Cyclic Redundancy Check
    """
    
    divisor = polynomial
    crc = data
    
    for s in range(size):
        if crc & (1 << (size - 1)):
            crc = (crc << 1) ^ divisor
        else:
            crc = crc << 1
    return crc & ((1 << size) - 1)
````

````python
def checksum(data, mod=256):
    """ 
    Function for checksum algorithm
    """
    return sum(data) % mod
````
